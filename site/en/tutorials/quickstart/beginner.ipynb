# Math in tensorFlow
1. tf.add() = tf.math.add()

  Returns x + y element-wise.

  **✅ tens + tens = good**  

  **✅notTens + notTons = good**

  **❎ notTens + tens = notgood (overFlow or underFlow)**
  ```
  tf.math.add(
      x, y, name=None
  )
  <!-- x and y is var in tensor  -->
  ```
  **example:**

  ```
  x = [1, 2, 3, 4, 5]
  y = 1
  tf.add(x, y)

  ```
2. tf.math.multiply() = tf.multiply()

  Returns x * y element-wise. ( ⚡not multiply matrix  )

  **✅ tens + tens = good**  

  **✅notTens + notTons = good**

  **❎ notTens + tens = notgood (overFlow or underFlow)**
  ```
 tf.math.multiply(
    x, y, name=None
)
  ```
  **example:**

  ```
  x = tf.constant(([1, 2, 3, 4]))
  tf.math.multiply(x, x)
  ```
3. subtract
  ```
  x = tf.constant([
    [12,3], [22,1], [11,33], [22,22], [55,33]])
   y = tf.constant([ [12,3], [22,1], [11,33], [22,22], [55,33]])
   tf.subtract(x,y)
  ```
4. divide
```
x = tf.constant([
    [12,3], [22,1], [11,33], [22,22], [55,33]])
```
