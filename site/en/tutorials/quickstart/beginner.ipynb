# Returns a tensor to shape of the input tensor.
```
tf.shape(
    input, out_type=None, name=None
)
like this
t = tf.constant([[[1, 1, 1], [2, 2, 2]], [[3, 3, 3], [4, 4, 4]]])
tf.shape(t)

```
# Returns the rank of a tensor.
```
tf.rank(
    input, name=None
)
like this
t = tf.constant([[[1, 1, 1], [2, 2, 2]], [[3, 3, 3], [4, 4, 4]]])
tf.rank(t) #=> 3
<!--  -->
t = tf.constant([1, 1, 1])
tf.rank(t) #=> 1
<!--  -->
t = tf.constant(5)
tf.rank(t) #=> 0
```
